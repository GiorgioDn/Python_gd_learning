```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#file upload
file_path = "california_housing_data.csv"
df = pd.read_csv(file_path)
```

The most relevant data are those returned by **std** and by comparing the maximum and minimum values, where the variation in values is most clearly visible. <br> <br>
**Specific Analysis**
- **MedInc**: Its mean is very close to the minimum value and far from the maximum. The standard deviation indicates a fairly consistent variation around the median.
- **HouseAge**: It has a mean that lies almost between the maximum and minimum values, so the data are very consistent; despite this, the standard deviation is quite high, meaning many data points will deviate significantly from the median.
- **AveRooms**: With a very low minimum and a very high maximum, the mean shows that most data points are very low, but they are much better concentrated around the median compared to HouseAge. Additionally, it may have some outliers, given the very high maximum.
- **AveBedrms**: The minimum is much lower than the maximum, but many values are similar, resulting in a very low standard deviation. Additionally, there may be some outliers, given the very high maximum.
- **Population**: The data values vary widely and deviate significantly from the standard deviation, resulting in a lack of homogeneity.
- **AveOccup**: It has a very high range of values between the maximum and minimum, and they are distributed unevenly given the high standard deviation. Additionally, it may have some outliers, given the very high maximum.
- **Latitude**: There is a very wide range between the maximum and minimum values; however, they are distributed unevenly due to the high standard deviation.
- **Longitude**: The values are fairly uniform, do not vary much from one another, and are distributed fairly close to the standard deviation.
- **MedHouseVal**: The values are fairly uniform, do not vary much from one another, and are distributed fairly close to the standard deviation. Despite this, they deviate significantly from the minimum.

```python
#Creation of the plot
plt.figure(figsize=(12,8))
#Create rows and columns within the figure, and then insert the charts at the corresponding index
plt.subplot(1,2,1)
sns.scatterplot(x=df.index, y=df["AveRooms"])
plt.title("AveRooms Distribution")
plt.xlabel("Index")
plt.ylabel("AveRooms")

plt.subplot(1,2,2)
sns.scatterplot(x=df.index, y=df["MedHouseVal"])
plt.title("MedHouseVal Distribution")
plt.xlabel("Index")
plt.ylabel("MedHouseVal")
#make sure the plots don't overlap
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(x=df["MedHouseVal"], y=df["AveRooms"])
plt.title("MedHouseVal and AveRooms Distribution")
plt.xlabel("MedHouseVal")
plt.ylabel("AveRooms")
plt.show()
```
