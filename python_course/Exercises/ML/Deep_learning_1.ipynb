```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
# 1. Create the empty container
model = models.Sequential(name="RealEstate_Network_v1")

# 2. Add the FIRST hidden layer
# It is essential to specify input_shape ONLY here.
# 32 neurons, ReLU activation for non-linearity.
model.add(layers.Dense(32, activation='relu', input_shape=(10,), name="Hidden_Layer_1"))

# 3. Add a SECOND hidden layer
# Note: Keras automatically understands that the input of this layer
# is the output of the previous one (32). No need to specify input_shape again.
model.add(layers.Dense(16, activation='relu', name="Hidden_Layer_2"))

# 4. Add the OUTPUT layer
# Only 1 neuron because we want a single number (price).
# No activation ('linear') because price can be any real value.
model.add(layers.Dense(1, activation='linear', name="Output_Layer"))

model.summary()
```
