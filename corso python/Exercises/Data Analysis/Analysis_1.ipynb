I dati più rilevanti sono quelli restituiti da **std** e confrontando il massimo ed il minimo, dove si può vedere meglio la variazione dei valori. <br> <br>
**Analisi specifica**
- **MedInc**: Ha una media che si avvicina molto al valore minimo e si distacca molto dal massimo. La deviazione standard ci dà una variazione intorno alla medianna abbastanza omogenea.
- **HouseAge**: Ha una media che si trova quasi tra il valore massimo ed il minimo, quindi i dati hanno valori molto omogenei, nonostante questo si ha una deviazione standard abbastanza elevata, quindi molti dati si discosteranno molto dalla mediana.
- **AveRooms**: Avendo un minimo molto basso ed un massimo molto alto, dalla media si può vedere che la maggior parte dei dati sono molto bassi, però si concentrano molto meglio sulla mediana rispetto all'HouseAge. Inoltre potrebbe avere qualchi outlier, dato il massimo molto elevato.
- **AveBedrms**: Il minimo è molto più basso del massimo però molti parametri hanno valori simili restituendo una deviazione standard molto bassa. Inoltre potrebbe avere qualchi outlier, dato il massimo molto elevato.
- **Population**: I dati hanno valori molto variabili e si discostano molto dalla deviazione standard risultando poco omogenei.
- **AveOccup**: Ha una variazione di valori molto alta tra il massimo ed il minimo e sono distribuiti in maniera poco omogenea dato il valore alto della deviazione standard. Inoltre potrebbe avere qualchi outlier, dato il massimo molto elevato.
- **Latitude**: Ha una variazione di valori molto alta tra il massimo ed il minimo, nonostante questo sono distribuiti in maniera poco omogenea dato il valore alto della deviazione standard.
- **Longitude**: Si hanno valori abbastanza omogenei che non si discostano molto tra di loro e sono distribuiti abbastanza vicino alla deviazione standard.
- **MedHouseVal**: Si hanno valori abbastanza omogenei che non si discostano molto tra di loro e sono distribuiti abbastanza vicino alla deviazione standard. Nonostante questo si discostano molto dal minimo

```python
#crazione del grafico
plt.figure(figsize=(12,8))
#crea all'interno della figure le righe e colonne per poi iserirci i grafici al corrispettivo indice
plt.subplot(1,2,1)
sns.scatterplot(x=df.index, y=df["AveRooms"])
plt.title("Distribuzione AveRooms")
plt.xlabel("Index")
plt.ylabel("AveRooms")

plt.subplot(1,2,2)
sns.scatterplot(x=df.index, y=df["MedHouseVal"])
plt.title("Distribuzione MedHouseVal")
plt.xlabel("Index")
plt.ylabel("MedHouseVal")
#fa in modo che non si sovrappongano i grafici
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(x=df["MedHouseVal"], y=df["AveRooms"])
plt.title("Distribuzione MedHouseVal e AveRooms")
plt.xlabel("MedHouseVal")
plt.ylabel("AveRooms")
plt.show()
```
